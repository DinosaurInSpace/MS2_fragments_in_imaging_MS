#wholebody_imaging_MS_dataset_10_region_metlin_library

To provide training data for the word2vec imaging MS database comparison, dataset #10 from the wholebody IMS project in METASPACE was processed.  First each assigned ion from core_metabolome_v3 was assigned to a region.  Next, prominent MS/MS peaks with assigned formulas from Metlin determined.  

Here, this datset will transformed into a valid input database for METASPACE.  This database of fragments can be used to search against the same dataset, and see if known fragments, derivied from neutral losses or in-source processes, are observed in IMS datsets.

https://drive.google.com/file/d/1se-HRQYLaH-jNTPru--BaKsGKa0cRe-c/view?usp=sharing

```python
path = '/Users/dis/PycharmProjects/word2vec/whole_body_imaging_core_metabolome_v3_f2.tsv'
raw_df = pd.read_csv(path, sep='\t')
```

```python
raw_df.head(5)
```

```python
# Prepare dataframe
df = raw_df.copy(deep=True)
df = df[df.metlin != 'no']
df = df.drop(columns=['metlin', 'm1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8'])
df = df.iloc[:-1,:]
df['name'] = df['formula'] + '_' + df['adduct'] + '_' + df['mz'] + '_' + df['moleculeNames']
df = df.drop(columns=['formula', 'adduct', 'mz', 'moleculeNames'])
df = df[['moleculeIds', 'name', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8']]
df = df.replace('Ornithine', 'C5H10N2O2+')
```

```python
# Process dataframe
out_df = pd.DataFrame()

fs = ['f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8']
for f in fs:
    temp_df = df[['moleculeIds', 'name', f]].copy(deep=True)
    temp_df = temp_df.dropna()
    temp_df['moleculeIds'] = temp_df['moleculeIds'] + '_' + f
    temp_df['name'] = temp_df['name'] + '_' + f
    temp_df = temp_df.rename(columns={f: 'formula'})
    out_df = pd.concat([out_df, temp_df], sort=True)

out_df['inchi'] = 'InChI=1S/H2O/h1H2'
out_df[['formula', 'adduct']] = out_df.formula.str.split('+', expand=True)
out_df = out_df.rename(columns={'moleculeIds': 'ids'})
```

```python
# Export dataframe
out_df.to_pickle('wholebody_imaging_MS_dataset_10_region_metlin_library.pickle')
out_df = out_df.iloc[:,0:4]
out_df.to_csv('wholebody_imaging_MS_dataset_10_region_metlin_library.csv', sep='\t')
```
